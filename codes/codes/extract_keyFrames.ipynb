```python
import cv2#importing cv2 library
import numpy as np#importing numpy library
import pydicom as dicom
import os
from os.path import join
import cv2
import shutil
from pathlib import Path
import itertools
```

```python
def score(image):

    taille=image.size
    sc=np.sum(image)/taille

    return sc
```

```python
def saveKeyFrames(listFrames,pathToSave):

    if os.path.exists(Path(pathToSave)) :
        shutil.rmtree(pathToSave)
    os.mkdir(pathToSave)

    for elet in listFrames:
        nameImage_to_write = join(pathToSave,elet[2])
        cv2.imwrite(nameImage_to_write,elet[1])
```

```python
os.chdir("path/to/datset/videos")
path=os.getcwd()
```

```python
listPatients=os.listdir()
```

```python
for unPatient in listPatients:
    pathPatient=join(path,unPatient)
    listCoro=os.listdir(unPatient)

    for unCoro in listCoro:
        pathCoro=join(pathPatient,unCoro)
        contenuCoro=os.listdir(pathCoro)

        for elet in contenuCoro:

            if elet=="frames" :
                pathFrames=join(pathCoro,elet)

                lesFrames=os.listdir(pathFrames)

                listFrames=[]

                for nameImg in lesFrames:

                    pathImg=join(pathFrames,nameImg)
                    im=cv2.imread(pathImg)

                    grayed = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

                    sc=score(grayed)

                    listFrames.append([sc,grayed,nameImg])
                print(len(listFrames))

                listFrames.sort(key=lambda x: x[0])
                keyFrames=listFrames[0:6]

                pathToSave=join(pathCoro,"keyFrames")
                saveKeyFrames(keyFrames,pathToSave)
```
